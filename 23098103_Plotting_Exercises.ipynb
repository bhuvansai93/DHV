## Problem:1
### Exponential Distribution
Histogram with Probability Density Function(PDF) line plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate Data from an Exponential Distribution
np.random.seed(40)  # For reproducibility

# Exponential distribution with mean 2 and random data with 450 size
p1_data = np.random.exponential(scale=2, size=450)  

# Calculate Mean and Median
mean = np.mean(p1_data)
median = np.median(p1_data)

# Plot Histogram
plt.figure(figsize=(10, 6),layout='tight')
plt.hist(p1_data, bins=20, color='mediumturquoise', edgecolor='black', alpha=0.7, density=True,zorder=3)

# Customize x and y axis ranges
plt.xlim(0,10)  # Focus on the main range of the exponential data
plt.ylim(0, 0.5)

# Set labels and title
plt.xlabel('Value', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.title('Histogram with PDF Line Plot (Exponential Distribution)', fontsize=16)

# Set x and y tick label sizes
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Calculate the PDF for the Exponential Distribution
lambda_param = 1 / 2  # Rate parameter for scale=2
x_values = np.linspace(0, 10, 1000)
pdf_values = lambda_param * np.exp(-lambda_param * x_values)

# Plot the PDF as a line
plt.plot(x_values, pdf_values, color='black', linewidth=2, label='Exponential PDF',zorder=3)

# Add Mean and Median Lines
plt.axvline(mean, color='green', linestyle='-.',alpha=0.5, linewidth=2, label=f'Mean: {mean:.2f}',zorder=3)
plt.axvline(median, color='purple', linestyle=':',alpha=0.5, linewidth=2, label=f'Median: {median:.2f}',zorder=3)

# Annotate Mean and Median
plt.text(mean+0.1, 0.35,f'Mean = {mean:.2f}',fontsize=14,color='green')
plt.text(median+0.1, 0.4,f'Median = {median:.2f}',fontsize=14,color='purple',zorder=3)

# Add legend for mean, median and PDF line
plt.legend(fontsize=14)
plt.grid(alpha=0.2)
# Display the plot
plt.show()
```

## Plots

1. Pairplots

### 1. Pairplots

```python
# Create pairplot function
def p2_pairplot():

    # for pairplots
    sns.pairplot(p2,hue="class", diag_kind="kde",palette='cubehelix')
    # add supertitle
    plt.suptitle("Pairplot of Variables by Class", y=1.05,fontsize=16)

    # Customize plot
    plt.ylabel("Variables", fontsize=14)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)

    # Show plot
    plt.show()
```

```python
# Call the pairplot function
p2_pairplot()
```

### 2. Correlation Heatmap

```python
# Create correlation matrix function
def p2_corrMatrix():
    # set figure size
    plt.figure(figsize=(8, 6))

    # heatmap for all numeric columns in dataset using seaborn
    sns.heatmap(p2.corr(), annot=True, cmap="coolwarm", fmt=".2f")

    # Customize plot
    plt.title("Correlation Heatmap",fontsize=16)
    plt.tick_params(labelsize=14)

    # show plot
    plt.show()

```

```python
# Call the correlation matrix function
p2_corrMatrix()
```

### 3. Boxplot

```python
# Create boxplot function
def p2_boxplot():
    
    # adjust figure size
    plt.figure(figsize=(10, 6))
    
    # Iterate all columns
    for i, column in enumerate(["var1", "var2", "var3"], 1):
        #subplots 1 row, 3 columns
        plt.subplot(1, 3, i)
        
        #plot boxplot using seaborn
        sns.boxplot(x="class", y=column, data=p2,hue='class')

        #customize the plot
        plt.title(f"Boxplot of '{column}' by Class",fontsize=16)
        plt.xlabel("Class", fontsize=14)
        plt.ylabel("Value", fontsize=14)
        plt.xticks(fontsize=14)
        plt.yticks(fontsize=14)

    plt.tight_layout()
    # Display plot
    plt.show()
```

```python
# Call the boxplot function
p2_boxplot()
```

### 4. Histogram and KDE Plot

```python
# Create kde histogram function
def p2_kde_hist():
    # 4. Histograms with KDE
    plt.figure(figsize=(12, 6))

    # Iterate all three variables
    for i, column in enumerate(["var1", "var2", "var3"], 1):
        # add 1 row with 3 columns subplots
        plt.subplot(1, 3, i)
        #plot histogram using seaborn
        sns.histplot(data=p2, x=column, hue="class", kde=True, palette="deep")

        # Customize plot
        plt.title(f"Histogram and KDE of {column}",fontsize=14)
        plt.xlabel(f'{column}',fontsize=14)
        plt.ylabel('Count',fontsize=14)
        plt.xticks(fontsize=14)
        plt.yticks(fontsize=14)
    
    plt.tight_layout()
    # Show plot
    plt.show()

```

```python
# Call the kde histogram function
p2_kde_hist()
```

### 5. Bar Chart Plot

```python
# Create Bar Chart Function
def p2_bar():
    
    # calculate means for each 'variable' as per 'class'
    p2_means = p2.groupby('class')[['var1', 'var2', 'var3']].mean().reset_index()
    
    #  add bar width and positions
    bar_width = 0.25
    x = np.arange(len(p2_means['class']))
    
    # Plot each variable
    plt.figure(figsize=(6, 5))
    plt.bar(x - bar_width, p2_means['var1'], width=bar_width, label='Var1', color='#1f77b4')
    plt.bar(x, p2_means['var2'], width=bar_width, label='Var2', color='#ff7f0e')
    plt.bar(x + bar_width, p2_means['var3'], width=bar_width, label='Var3', color='#2ca02c')
    
    # Customize the plot
    plt.title("Mean of Variables by Class", fontsize=16)
    plt.xlabel("Class", fontsize=14)
    plt.ylabel("Mean Value", fontsize=14)
    plt.xticks(ticks=x, labels= p2_means['class'], fontsize=14)
    plt.yticks(fontsize=14)
    plt.legend(title="Variables", fontsize=14)
    
    # Display the plot
    plt.tight_layout()
    plt.show()
```

```python
# Call the bar chart function

p2_bar()
```
